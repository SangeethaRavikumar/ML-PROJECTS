```python
df["online_order"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("online_order",axis = 1)
y = df["online_order"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python

lr = LogisticRegression()
start = time.time()
lr.fit(X_train,y_train)
end = time.time()
print("Time Taken:",end-start)
```

```python
y_pred = lr.predict(X_test)
print("Acurracy Score:",accuracy_score(y_test,y_pred))
print("Precision Score:",precision_score(y_test,y_pred))
print("Recall Score:",recall_score(y_test,y_pred))
print("f1 score    :",f1_score(y_test,y_pred))
```

# Decision Tree Clasifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
dtc = DecisionTreeClassifier()
```

```python
param_depth = {"max_depth":[3,4,5,6,7,8],
              "min_samples_split":[2,3,4],
              "min_samples_leaf":[1,2,3,4]
              }
grid_search = GridSearchCV(dtc,param_depth,cv = 5)
grid_search.fit(X_train,y_train)
print(grid_search.best_params_)
```

```python
dtc = DecisionTreeClassifier(random_state = 0,max_depth = 8,min_samples_leaf = 1,min_samples_split = 2)
start = time.time()
dtc.fit(X_train,y_train)
end = time.time()
print("Time Taken:",end-start)
```

```python
y_pred = dtc.predict(X_test)
print("Acurracy Score:",accuracy_score(y_test,y_pred))
print("Precision Score:",precision_score(y_test,y_pred))
print("Recall Score:",recall_score(y_test,y_pred))
print("f1 score    :",f1_score(y_test,y_pred))
```
